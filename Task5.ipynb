```python
import numpy as np
import pandas as pd
import warnings
```

##**pandas: Remove NaN (missing values) with dropna()** <br>
##**pandas: Replace NaN (missing values) with fillna()**<br>
##**pandas: Detect and count NaN (missing values) with isnull(), isna()**

```python
data = pd.DataFrame({'col1': [0, np.nan, np.nan, 3, 4],
                   'col2': [np.nan, 1, 2, np.nan, np.nan],
                   'col3': [4, np.nan, np.nan, 7, 10]})
data
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
data.interpolate()
```

### Row or column
### The same value is repeated for the rightmost NaN
### while the leftmost NaN remains unchanged.


```python
data.interpolate(axis=1)
```

#limit argument

*   ### it specify the maximum number of consecutive NaN values to interpolate.

```python
data.interpolate(limit=1)
```

## **Direction to interpolate : limit direction**

```python
forward = data.interpolate(limit=1,limit_direction='forward')
forward
```

```python
backward = data.interpolate(limit=1,limit_direction='backward')
backward
```

```python
both = data.interpolate(limit=1,limit_direction='both')
both
```

##**Interpolate or extrapolate or both: limit_area**
*   ### inside
*   ### outside
*   ### both



```python
inside = data.interpolate(limit=1,limit_area='inside')
inside
```

```python
outside = data.interpolate(limit=1,limit_area='outside')
outside
```

```python
both1 = data.interpolate(limit_direction='both',limit_area='outside',)
both1
```

## Operator in place

*   ### can directly modify the original object by setting inplace = true



```python
data.interpolate(inplace=True)
data
```

## **linear interpolation**

```python
data2 = pd.Series([0, np.nan, np.nan, 3],
              index=[0, 4, 6, 8])

a = data2
b = data2.interpolate()
c = data2.interpolate('index')
print(a)
print(b)
print(c)
```

**Using existing values:**
*   ffill
*   pad
*   bfill
*   backfill

```python
data3 = pd.Series([np.nan, 1, np.nan, 2, np.nan])
data3
```

**ffill forward fill**

```python
a= data3.interpolate("ffill")
a
```

**bfill backward fill**

```python
b=data3.interpolate('bfill')
b
```

**ffill and bfill using method()**

```python
#ffill forward fill
a = data3.fillna(method='ffill')
print(a)

print()
#bfill backward fill
b = data3.fillna(method='bfill')
print(b)
```

**Spline Interpolation**

```python
spline = pd.Series([0, 10, np.nan, np.nan, 4, np.nan],
              index=[0, 2, 5, 6, 8, 12])

spline.interpolate('spline', order=2)
```

```python
spline.index = range(6)
spline
```

**time series data**
