```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mat
import seaborn as see
import sklearn.linear_model as lm
lreg = lm.LinearRegression()
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = dta.year.values
x = x.reshape(47,1)
x.shape
```

```python
y = dta.pci.values
y
```

# **Simple Linear regrssion**

```python
lreg.fit(x,y)
```

```python
dta["Predicted_val"] = lreg.predict(x)
dta
```

```python
predicted = lreg.predict(x)
predicted
```

```python
from sklearn.metrics import r2_score
```

```python
accu = r2_score(y,predicted)
print("Accuracy of the model is ",accu*100)
```

```python
mat.plot(x,y,color='red',label='Actual PCI')
mat.scatter(x,y,color='red')

mat.plot(x,predicted,color='green',label='Predicted PCI')
#mat.scatter(x,predicted,color='green')
mat.title('Canada PCI Over the Years')
mat.xlabel('Year')
mat.ylabel('PCI (Per Capita Income)')
mat.legend()
mat.show()
```

# **Polynomial Regression**

```python
from sklearn.preprocessing import PolynomialFeatures

# Degree 2
preg1 = PolynomialFeatures(degree=2)
xpoly2 = preg1.fit_transform(x)
```

```python
lreg.fit(xpoly2,y)
```

```python
prevalue = lreg.predict(xpoly2)
mat.scatter(x,y,color='blue')
mat.plot(x,prevalue,color='green',label='Regression line')
mat.title('Canada PCI Over the Year with degree=2')
mat.xlabel('Year')
mat.ylabel('PCI (Per capita Income)')
mat.legend()
mat.show()
```

degree=3

```python
preg3 = PolynomialFeatures(degree=3)
xpoly3 = preg3.fit_transform(x)
```

```python
lreg.fit(xpoly3,y)
```

```python
prevalue = lreg.predict(xpoly3)
mat.scatter(x,y,color='blue')
mat.plot(x,prevalue,color='green',label='Regression line')
mat.title('Canada PCI Over the Year with degree=3')
mat.xlabel('Year')
mat.ylabel('PCI (Per Capita Income)')
mat.legend()
mat.show()
```

Degree=4

```python
preg4 = PolynomialFeatures(degree=4)
xpoly4 = preg4.fit_transform(x)
```

```python
lreg.fit(xpoly4,y)
```

```python
prevalue = lreg.predict(xpoly4)
mat.scatter(x,y,color='brown')
mat.plot(x,prevalue,color='green',label='Regression line')
mat.title('Canada PCI Over the Year with degree=4')
mat.xlabel('Year')
mat.ylabel('PCI (Per Capita Income)')
mat.legend()
mat.show()
```
